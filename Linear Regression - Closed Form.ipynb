```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X = np.linspace(0, 100, 200)
delta = np.random.uniform(-15, 10, X.size)
Y = 0.4 * X + 3 + delta

plt.figure(figsize=(12, 9))
plt.scatter(X, Y)
plt.show()
```

We can get the true line of best fit with the best weight and bias using the closed form linear regression formula:

Closed form Linear Regression to find the weights and biases: $W=(X^TX)^{-1}X^TY$

We have to add a column of ones to X to get the correct bias. Once put through the closed form equation, we get 2 values: the weight and bias.

```python
class LinearRegressionClosedFormModel:
    """
    This is a custom Linear Regression Model that uses the closed-form equation to find the best possible weight and bias.
    We can use the above equation, and add a column of ones to the end to get the bias term.
    """

    def __init__(self):
        self.w = 1
        self.b = 1

    def fit(self, X, Y):
        # Add a column of ones to X for the bias
        X_ones = np.c_[X, np.ones(X.shape[0])]

        print(f"Shape of X_ones: {X_ones.shape}")

        # Closed Form Linear Regression Equation
        w_and_b = np.linalg.inv(X_ones.T @ X_ones) @ X_ones.T @ Y

        self.w, self.b = w_and_b[0], w_and_b[1]

        print(f"Weight: {self.w} | Bias: {self.b}")

    def predict(self, X):
        return self.w * X + self.b
```

```python
model = LinearRegressionClosedFormModel()
model.fit(X, Y)
Y_pred = model.predict(X)

plt.figure(figsize=(12, 9))
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color="red")
plt.show()
```
